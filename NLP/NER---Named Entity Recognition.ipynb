```python
import spacy
from spacy import displacy
NER = spacy.load("en_core_web_sm")
```

```python
text = """Apple Inc. is planning to launch a new product next month. The CEO, Tim Cook, 
announced during the company's annual conference that they will unveil the latest iPhone model. 
The event is scheduled to take place at the Steve Jobs Theater in Cupertino, California. 
Analysts predict that the new iPhone will feature advanced technology and improved performance. Apple's stock price surged after the announcement, reaching an all-time high. The company continues to innovate and dominate the technology market with its cutting-edge products.
"""
```

```python
text1 = NER(text)
```

```python
for word in text1.ents:
    print(word.text, word.label_)
```

```python
spacy.explain("ORG")
```

```python
spacy.explain("PERSON")
```

```python
spacy.explain("GPE")
```

```python
displacy.render(text1)
```

```python
displacy.render(text1, style="ent", jupyter=True)
```
